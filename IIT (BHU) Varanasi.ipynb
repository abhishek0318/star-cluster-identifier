# Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
import numpy as np
```

# Loading data

```python
filename = 'Dataset.txt'
column_names = ['X stellar position in a reference system where the cluster center', 'Y stellar position in a reference system where the cluster center', 'F275W calibrated magnitude', 'F275W photometric RMS', 'F275W quality-fit parameter', 'F275W sharp parameter', 'Number of F275W exposures the source is found in', 'Number of F275W exposures the source is well measured', 'F336W calibrated magnitude', 'F336W photometric RMS', 'F336W quality-fit parameter', 'F336W sharp parameter', 'Number of F336W exposures the source is found in', 'Number of F336W exposures the source is well measured', 'F438W calibrated magnitude', 'F438W photometric RMS', 'F438W quality-fit parameter', 'F438W sharp parameter', 'Number of F438W exposures the source is found in', 'Number of F438W exposures the source is well measured', 'F606W calibrated magnitude', 'F606W photometric RMS', 'F606W quality-fit parameter', 'F606W sharp parameter', 'Number of F606W exposures the source is found in', 'Number of F606W exposures the source is well measured', 'F814W calibrated magnitude', 'F814W photometric RMS', 'F814W quality-fit parameter', 'F814W sharp parameter', 'Number of F814W exposures the source is found in', 'Number of F814W exposures the source is well measured [99: saturated star]', 'Membership Probability', 'Right ascension of the star', 'Declination of the star', 'Identification number of the star', 'Iteration the star was found']
data = pd.read_table(filename, skiprows=43, header=None, delim_whitespace=True, names=column_names)
data.head()
```

```python
data.describe()
```

```python
data.isnull().values.any()
```

We observe that the dataset has no missing values, therefore we can proceed with our analysis.  
As given in the problem statement, we ignore stars fainter than $m_{F606W}=22$

```python
rows_before = data.shape[0]
data = data[data['F606W calibrated magnitude'] <= 22]
rows_after = data.shape[0]
print(rows_after/rows_before)
```

This operation removes approximately 2/3rds of the stars.  
We remove stars with membership probability less than 90%, and some outliers to draw the CMD diagram.

```python
data1 = data[data['Membership Probability'] > 90] # stars with membership probability greater than 90%
data1 = data1[((data1['F438W calibrated magnitude'] - data1['F606W calibrated magnitude']).abs() < 20) & (data1['F606W calibrated magnitude'] > 0)] # Remove outliers
```

```python
plt.xlabel('Right ascension of the star')
plt.ylabel('Declination of the star')
plt.title('Globular Cluster')
plt.scatter(data1['Right ascension of the star'], data1['Declination of the star'], s=1, edgecolors='none', c='k')
plt.show()
```

```python
plt.figure(figsize=(20,20))
plt.gca().invert_yaxis()
plt.xlabel('B-V')
plt.ylabel('V')
plt.title('Colour Magnitude Diagram')
plt.scatter(data1['F438W calibrated magnitude'] - data1['F606W calibrated magnitude'], data1['F606W calibrated magnitude'], s=1, edgecolors='none', c='k')
```

```python
right_ascension_center = data1['Right ascension of the star'].mean() # Assuming
right_ascension_diff = data1['Right ascension of the star'] - right_ascension_center

declination_center = data1['Declination of the star'].mean() # Assuming
declination_diff = data1['Declination of the star'] - declination_center

distance_center = (right_ascension_diff ** 2 + declination_diff ** 2) ** (1/2)
print("Right ascension center: ", right_ascension_center, ", Declination Center: ", declination_center, sep='')
```

```python
def luminosity_sum(data):
    return np.power(10, -(data['F606W calibrated magnitude'] + 26.916) / 2.5).sum()
    
```

```python
data1['Distance from center'] = distance_center
```

```python
luminosity_sum(data1.sort_values(by=['Distance from center'], axis=0)[:24273]) / luminosity_sum(data1)
```

```python
right_ascension_half = 205.56 - right_ascension_center
declination_center_half = 28.3802 - declination_center
distance_half = ((right_ascension_half ** 2) + (declination_center_half ** 2)) ** (1/2)
print(right_ascension_half, declination_center_half, distance_half)
```

```python
plt.xlabel('Right ascension of the star')
plt.ylabel('Declination of the star')
plt.title('Globular Cluster with half light radius')
plt.scatter(data1['Right ascension of the star'], data1['Declination of the star'], s=1, edgecolors='none', c='k')
circle1 = plt.Circle((right_ascension_center, declination_center), distance_half, color='r', fill=False, lw=2)
plt.gcf().gca().add_artist(circle1)
plt.show()
```

```python
def alpha(m):
    if m <= 0.08:
        return 0.3
    elif m <= 0.5:
        return 1.3
    else:
        return 2.3
```

```python
s =  0
for m in np.linspace(0.01, 0.8, 80):
    s += m ** (-alpha(m)) * 0.01
print(s)
```

```python
n = data1.shape[0]
xi = n / s
print(xi)
```
